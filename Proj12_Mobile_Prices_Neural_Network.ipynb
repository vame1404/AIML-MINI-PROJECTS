```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("mobile_prices.csv")
X = df.iloc[:,:20].values
y = df.iloc[:,20:21].values
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.1)
oh = OneHotEncoder()
ytrain = oh.fit_transform(ytrain).toarray()
```

```python
model=Sequential()
model.add(Dense(16,input_dim=20,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(4,activation="softmax")) 

#if you have a multiclass classification the activation function is doftmax
#for binary class clssification put sigmoid as classification activation

model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])

#when multiclass classification categorical_crossentropy 
#for binary class classification use binary_crossentropy 
```

```python
history = model.fit(Xtrain, ytrain, epochs=100, batch_size=64)
ypred = model.predict(Xtest)
ypred = np.argmax(ypred,axis=1)

#probability distribution for y pred
# (it will choose 3 cause hihghest probability out of all others for 3 is 0.5)
# for 0  probability is 0.1 
# for 1 is probability 0.15 
# for 2 is probability 0.25 
# for 3 is probability 0.5
```

```python
score = accuracy_score(ypred,ytest)
print('Accuracy score is',100*score,'%')
```
