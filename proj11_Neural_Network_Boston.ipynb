```python
from sklearn import datasets
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
# boston = datasets.load_boston()
# X=boston.data
# Y=boston.target
```

```python
Bosdata=pd.read_csv('BostonHousing.csv')
X=Bosdata.iloc[:,0:13]
y= Bosdata.iloc[:,13] #MEDV:median value of owner-occuied home in $1000
```

```python
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
```

```python
model= Sequential()
#copy paste these 2 lines to add more
# layers but chnage the nummber of neurons here it is 13 dense(13,...)
model.add(Dense(13,input_dim=13,activation="relu"))
model.add(Dense(1))  
#adam is gradient decent advance version
# optimizer gradient decent can also be used
# model.add(Dense(10,input_dim=13,activation="relu"))
# model.add(Dense(2)) 
model.compile(loss="mean_squared_error",optimizer="adam",metrics=["accuracy"])
#accuracy is not specified here yet
```

```python
#weigths and biases are trained in this
history=model.fit(Xtrain,ytrain,epochs=150,batch_size=10)

ypred=model.predict(Xtest)
ypred=ypred[:,0]
```

```python
error=np.sum(np.abs(ytest-ypred))/np.sum(np.abs(ytest))*100
print('Prediction Error is',error,'%')
```
